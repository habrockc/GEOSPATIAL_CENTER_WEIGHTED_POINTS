```python
# -*- coding: utf-8 -*-
"""
Created on Mon Apr 27 11:48:48 2020

@author: cdhabro
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
from shapely.geometry import Point, LineString
import contextily as ctx
from descartes import PolygonPatch
%matplotlib inline

"""
WGS 84
EPSG:4326

NAD83(HARN) / Texas South Central (ftUS)
EPSG:2919

# try not to use if can avoid
WGS 84 / Pseudo-Mercator -- Spherical Mercator, Google Maps, OpenStreetMap, Bing, ArcGIS, ESRI
EPSG:3857
"""

### coordinate data ###
#chris Dairy Ashford & W. Airport
#abby Wilcrest & Memorial
#amber Fry & Kingsland
#molly Sprouts on Yale in Heights
#greg White Oak & Studewood
#josh Cactus Cove Bar & Grill
#fatima The Galleria

# create our data using a dictionary holding names, longitude, latitude, and household count
personnelCoords = {"Chris":[-95.597757, 29.648080, 3],
                   "Abby":[-95.575944, 29.772602, 4],
                   "Amber":[-95.718956, 29.77525, 2],
                   "Molly":[-95.400196, 29.774792, 1],
                   "Greg":[-95.387697, 29.781622, 2],
                   "Josh":[-95.436674, 29.788920, 1],
                   "Fatima":[-95.463633, 29.739631, 1]
                   }

# create function to convert dataframe to geodataframe
def df_to_points_gdf(dataset, longitude, latitude, inputCRS="EPSG:4326"):
    data = dataset
    lon = data[longitude]
    lat = data[latitude]
    inCRS = inputCRS
    geomData = [Point(xy) for xy in zip(lon, lat)]
    newGeoDataFrame = gpd.GeoDataFrame(data, crs=inCRS, geometry=geomData)
    data["CRS_source"] = inCRS
    return newGeoDataFrame

# create dataframe with lon/lats
df = pd.DataFrame.from_dict(personnelCoords,orient='index', columns=["LON","LAT", "HOUSEHOLD_COUNT"])
df.index.name = "NAME"
df.reset_index(inplace=True)

# set random seed and create random integer for each row (not needed but did so for testing purposes)
np.random.seed(7)
df["RANDOM"] = np.random.randint(1,21, size=df.shape[0])

# create variable of raw CRS
coordRefSys = "EPSG:4326"

# use function to convert dataframe to geodataframe with raw CRS
gdf = df_to_points_gdf(df, "LON", "LAT")

# project raw CRS to local 2-d cartesian projection for spatial analysis
gdf.to_crs("EPSG:2919", inplace=True)
gdf["CRS_projection"] = gdf.crs

# create variable to hold the geodataframe's newly projected CRS
getGDFCRS = gdf.crs

# now that data is projected from spherical coordinates to 2-D Cartesian planar geometry points
# we can calculate the mean X and mean Y to obtain the geometric center
xMean = np.mean(gdf.geometry.x)
yMean = np.mean(gdf.geometry.y)
centroid = [[xMean, yMean], ]

# create variable holding the 'weight' column
weightColumn = gdf["HOUSEHOLD_COUNT"]

# same as mean center but obtain weighted values (sum(all 'X' * weight) & divide by sum(all weights))
# and then do the same for 'Y'
xMeanWeighted = sum(gdf.geometry.x * weightColumn) / sum(weightColumn)
yMeanWeighted = sum(gdf.geometry.y * weightColumn) / sum(weightColumn)
centroidWeighted = [[xMeanWeighted, yMeanWeighted], ]

# obtain distance between mean center & weighted mean center
meanWeightedMeanLine = LineString([(xMean, yMean), (xMeanWeighted, yMeanWeighted)])
lineDistance = meanWeightedMeanLine.length
# lineDistanceMiles = lineDistance / 5280

# create dataframes for mean center & weighted mean center
dfCentroid = pd.DataFrame(centroid, columns=["LONcentroid", "LATcentroid"])
dfCentroidWeighted = pd.DataFrame(centroidWeighted, columns=["LONcentroidWeighted", "LATcentroidWeighted"])

# convert each new dataframe into a geodataframe from previously created function and
# because the center & weighted center were obtained from projected, pass in the projection
# as the inputCRS
gdfCentroid = df_to_points_gdf(dfCentroid, "LONcentroid", "LATcentroid", inputCRS=getGDFCRS)
gdfCentroidWeighted = df_to_points_gdf(dfCentroidWeighted, "LONcentroidWeighted", "LATcentroidWeighted", inputCRS=getGDFCRS)


# get specific row to use to calculate distance against all other points
gdfCentroidOnly = gdfCentroid.iloc[0,:]
gdfCentroidWeightedOnly = gdfCentroidWeighted.iloc[0,:]

# calculate distances (in projection units, this case is feet), then convert to miles
gdf["DIST_CENTER_FEET"] = gdf.geometry.distance(gdfCentroidOnly.geometry)
gdf["DIST_CENTER_WGT_FEET"] = gdf.geometry.distance(gdfCentroidWeightedOnly.geometry)
gdf["DIST_CENTER_MILE"] = gdf["DIST_CENTER_FEET"] / 5280
gdf["DIST_CENTER_WGT_MILE"] = gdf["DIST_CENTER_WGT_FEET"] / 5280
gdf["CENTER_VS_WGT_CENTER_DIFF_MILE"] = gdf["DIST_CENTER_WGT_MILE"] - gdf["DIST_CENTER_MILE"]
gdf["WGT_CENTER_CLOSER_OR_FARTHER"] = np.where(gdf["CENTER_VS_WGT_CENTER_DIFF_MILE"] <0, "closer", "farther")

# create list of all geodataframes to plot and iterate each while 'unprojecting', so
# specifically, reproject to a CRS suitable for most web maps (in this case, EPSG:4326 (WGS 84))
listToUnproject = [gdf, gdfCentroid, gdfCentroidWeighted]
for each in listToUnproject:
    print("Current CRS:", each.crs)
    each.to_crs(coordRefSys, inplace=True)
    #each.to_crs("EPSG:3857", inplace=True)
    #each.to_crs("EPSG:4326", inplace=True)
    print("Newly converted CRS:", each.crs)


### construct polygon from points by converting all points into a single MultiPoint type
### then get the geometry, then create a convex_hull, then add via PolygonPatch module
### this part can be commented out to remove shapes and stick to just points and centers
empPts = gdf.geometry
empPtsSingleMultiPoint = empPts.unary_union
empPtsConvexHull = empPtsSingleMultiPoint.convex_hull
empPtsEnvelope = empPtsSingleMultiPoint.envelope
#empPtsShape = PolygonPatch(empPtsConvexHull, ec="blue", alpha=0.5, zorder=2, fill=False)
empPtsShapeConvex = PolygonPatch(empPtsConvexHull, fc="yellow", alpha=0.2, zorder=2)
empPtsShapeEnvelope = PolygonPatch(empPtsEnvelope, ec="black", fill=False)
### end construct polygon from points


# create the plot and add each geodataframe to it along with other parameters
fig, ax = plt.subplots(1,1)


### if commenting out shapes, comment out ax.add_patch rows also
ax.add_patch(empPtsShapeConvex)
ax.add_patch(empPtsShapeEnvelope)
### end if comment out shapes, comment out ax.add_patch rows also

gdf.plot(ax=ax, color="red", figsize=(12, 12), marker="o", markersize=weightColumn*10)
gdfCentroid.plot(ax=ax, color="blue", marker="o", markersize=24)
gdfCentroidWeighted.plot(ax=ax, color="black", marker="x", markersize=36)
ctx.add_basemap(ax=ax, alpha=0.5, zoom=14, crs=coordRefSys)
plt.title("Abby's Team\n""Distance from geographic center to weighted geographic center:\n {:.0f} ft. ({:.2f} miles)".format(lineDistance, lineDistance/5280,2))
plt.xlabel("LONGITUDE")
plt.ylabel("LATITUDE")
#ax.axis("off")
```
